### BusinessTravel: Ordinal Encoding
### All others: Nominal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer 
from sklearn.compose import make_column_transformer
```

```python
ohe_encoder = OneHotEncoder()
ord_encoder = OrdinalEncoder()
```

```python
X.info()
```

```python
X.drop("OverTime",axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.30,
                                                    random_state=42,
                                                    stratify=y)  
```

```python
ct = make_column_transformer( 
            (ord_encoder,["BusinessTravel"]),
            (ohe_encoder, ["Department","EducationField", "Gender","JobRole", "MaritalStatus"]),
            remainder = 'passthrough')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

pipe = Pipeline([
    ('Column_Transformations', ct),
    ('Logistic Regression', LogisticRegression()),
])
```

```python
pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = pipe.predict(X_train)
accuracy_score(y_train, y_pred)
```

```python
from sklearn.metrics import accuracy_score
y_pred = pipe.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_auc_score
y_pred = pipe.predict_proba(X_train)[:,1]
roc_auc_score(y_train, y_pred)
```

```python
from sklearn.metrics import roc_auc_score
y_pred = pipe.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_pred)
```

```python
import pickle
with open ('./models/pipe.bin', 'wb' ) as f_out: 
    pickle.dump(pipe, f_out)
```
